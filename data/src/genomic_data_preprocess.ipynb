```python
import pandas as pd
import numpy as np
import os 
```

```python
cwd = os.getcwd()
```

```python
df_v = pd.read_csv(cwd + '/raw_data/SARS-CoV-2_Variant_Proportions.csv')
```

```python
df_v['week_ending'] = pd.to_datetime(df_v['week_ending'])
df_v['published_date'] = pd.to_datetime(df_v['published_date'])
df_v = df_v[(df_v['published_date'] >= '2022-01-01')&(df_v['time_interval'] == 'weekly')]
df_v = df_v.reset_index().drop(columns = ['index'])
df_v = df_v[df_v['usa_or_hhsregion'] != 'USA']
df_v = df_v.reset_index().drop(columns = ['index'])
```

```python
df_v = df_v[df_v['modeltype'] == 'smoothed'].drop(columns = ['nchs_or_count_flag'])
df_v = df_v.reset_index().drop(columns = ['index'])
df_v = df_v.drop(columns = ['share_hi', 'share_lo'])
```

```python
index = df_v[df_v['variant'].str[:4] == 'BA.2'].index
df_v.loc[index, 'variant'] = 'BA.2'

index = df_v[df_v['variant'].str[:4] == 'BA.4'].index
df_v.loc[index, 'variant'] = 'BA.4'

index = df_v[df_v['variant'].str[:3] == 'XBB'].index
df_v.loc[index, 'variant'] = 'XBB'

index = df_v[df_v['variant'].str[:4] == 'BA.5'].index
df_v.loc[index, 'variant'] = 'BA.5'

index = df_v[df_v['variant'].str[:4] == 'BQ.1'].index
df_v.loc[index, 'variant'] = 'BQ.1'

index = df_v[df_v['variant'].str[:3] == 'B.1'].index
df_v.loc[index, 'variant'] = 'B.1'

df_v = df_v.groupby(['usa_or_hhsregion', 'week_ending', 'variant', 'modeltype',
             'time_interval', 'published_date']).sum().reset_index()
```

```python
df_v['share'] = round(df_v['share'], 3)
df_v = df_v[df_v['share'] > 0].reset_index().drop(columns = ['index'])
df_v = df_v.drop(columns = ['modeltype', 'time_interval'])
```

```python
df_v1 = pd.DataFrame(columns = ['usa_or_hhsregion', 'variant', 'published_date'])
```

```python
for date in df_v['published_date'].unique():
    for region in df_v['usa_or_hhsregion'].unique():
        test = df_v[(df_v['published_date'] == date)&\
        (df_v['usa_or_hhsregion'] == region)].sort_values('share', ascending = False).head(5).iloc[:, :-1]
        df_v1 = df_v1.append(test)
    
```

```python
for index, row in df_v1.iterrows():
    share = df_v[(df_v['published_date'] == row['published_date'])&\
                 (df_v['usa_or_hhsregion'] == row['usa_or_hhsregion'])&\
    (df_v['variant'] == row['variant'])]['share'][-2:].to_numpy()
    if len(share) >= 2:
        df_v1.loc[index, ['prev_2w', 'prev_1w']] = share
    
```

```python
df_v1 = df_v1.dropna().reset_index().drop(columns = ['index'])
df_v1 = df_v1[df_v1['prev_2w'] > 0.01].reset_index().drop(columns = ['index'])
```

```python
df_v1 = df_v1.drop(columns = ['week_ending']).drop_duplicates().reset_index().drop(columns = ['index'])
```

```python
df_v1.to_pickle(cwd + '/processed_data/top_5_variant.pkl')
```
